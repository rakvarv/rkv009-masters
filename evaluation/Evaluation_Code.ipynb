```python
# GPT3 LOCATION ACCURACY - NO ONTOLOGY SPECIFIED
# Initialize lists for actual and predicted locations
actual_locations = []
predicted_locations_gpt3 = []

# Read CSV file and process rows
with open('test_processedgpt3.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        actual_location_json = json.loads(row['actual'])
        predicted_location_json = json.loads(row['predictedgpt3'])
        
        # Extract locations from actual JSON
        actual_locations.extend([location['name'] for location in actual_location_json['locations']])
        
        # Extract locations from predicted GPT-3 JSON
        predicted_locations_gpt3.extend([location['name'] for location in predicted_location_json['locations']])

# Function to check if a location is a partial match
def is_partial_match(location, location_list):
    for actual_loc in location_list:
        if location in actual_loc or actual_loc in location:
            return True
    return False

# Calculate accuracy based on partial matches
total_predictions = len(predicted_locations_gpt3)
correct_predictions = sum(1 for predicted_loc in predicted_locations_gpt3 if is_partial_match(predicted_loc, actual_locations))
accuracy = correct_predictions / total_predictions if total_predictions > 0 else 0

print("Total predictions:", total_predictions)
print("Correct predictions:", correct_predictions)
print("Accuracy:", accuracy)
```

```python
# GPT3 RELATIONSHIP ACCURACY - NO ONTOLOGY SPECIFIED
# Initialize variables
total_relationships = 0
correct_relationships = 0

# Open the CSV file and iterate over rows
with open('test_processedgpt3.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        actual_relationships_json = json.loads(row['actual'])
        predicted_relationships_json = json.loads(row['predictedgpt3'])
        
        # Ensure 'relationships' key exists in predicted JSON
        if 'relationships' not in predicted_relationships_json:
            predicted_relationships_json['relationships'] = []
        
        # Extract relationships from actual JSON
        actual_relationships = actual_relationships_json.get('relationships', [])
        
        # Extract relationships from predicted GPT-3 JSON
        predicted_relationships = predicted_relationships_json.get('relationships', [])
        
        # Check for correct predictions
        for pred_rel in predicted_relationships:
            for act_rel in actual_relationships:
                # Check original order
                if (pred_rel['from'] == act_rel['from'] and
                    pred_rel['relation'] == act_rel['relation'] and
                    pred_rel['target'] == act_rel['target']):
                    correct_relationships += 1
                    break  # Move to the next predicted relationship
                
                # Check reversed order
                if (pred_rel['from'] == act_rel['target'] and
                    pred_rel['relation'] == act_rel['relation'] and
                    pred_rel['target'] == act_rel['from']):
                    correct_relationships += 1
                    break  # Move to the next predicted relationship
        
        # Increment total relationships count
        total_relationships += len(actual_relationships)

# Calculate accuracy
accuracy_relationships = correct_relationships / total_relationships if total_relationships > 0 else 0

# Print results
print("Total relationships:", total_relationships)
print("Correct relationships:", correct_relationships)
print("Accuracy:", accuracy_relationships)
```

```python
# Define the function to calculate the soft evaluation score
def calculate_soft_evaluation_score(actual_relationships_json, predicted_relationships_json):
    actual_relationships = actual_relationships_json.get('relationships', [])
    predicted_relationships = predicted_relationships_json.get('relationships', [])
    
    # Initialize a counter for correct relationships
    correct_relationships_count = 0
    
    # Loop through predicted relationships
    for pred_rel in predicted_relationships:
        # Check if any part of the predicted relationship matches with any part of the actual relationships
        for actual_rel in actual_relationships:
            # Count the number of matching parts
            match_count = 0
            if pred_rel['from'] == actual_rel['from'] or pred_rel['from'] == actual_rel['target']:
                match_count += 1
            if pred_rel['relation'] == actual_rel['relation']:
                match_count += 1
            if pred_rel['target'] == actual_rel['target'] or pred_rel['target'] == actual_rel['from']:
                match_count += 1
            # Check if at least two parts match
            if match_count >= 2:
                correct_relationships_count += 1
                break  # Break out of inner loop once a match is found
    
    # Calculate soft score based on the proportion of correct relationships
    soft_score = correct_relationships_count / len(actual_relationships) if len(actual_relationships) > 0 else 0
    
    return soft_score

# Initialize variables
total_relationships = 0
correct_relationships = 0
correct_soft_relationships = 0  # Initialize counter for correct soft evaluation scores
total_predicted_relationships = 0  # Initialize counter for total predicted relationships

# Initialize scores for hard and soft evaluation
hard_evaluation_score = 0
soft_evaluation_score = 0

# Open the CSV file and iterate over rows
with open('test_processedgpt3ontology.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        actual_relationships_json = json.loads(row['actual'])
        predicted_relationships_json = json.loads(row['predictedgpt3ont'])
        
        # Ensure 'relationships' key exists in predicted JSON
        if 'relationships' not in predicted_relationships_json:
            predicted_relationships_json['relationships'] = []
        
        # Extract relationships from actual JSON
        actual_relationships = actual_relationships_json.get('relationships', [])
        
        # Extract relationships from predicted GPT-3 JSON
        predicted_relationships = predicted_relationships_json.get('relationships', [])
        
        # Update scores for hard evaluation
        for relation in predicted_relationships:
            if relation in actual_relationships:
                correct_relationships += 1
                hard_evaluation_score += 1  # Increment score for correct prediction
        
        # Increment total relationships count
        total_relationships += len(actual_relationships)
        
        # Calculate soft evaluation score
        soft_evaluation_score += calculate_soft_evaluation_score(actual_relationships_json, predicted_relationships_json)
        
        # Increment total predicted relationships count
        total_predicted_relationships += len(predicted_relationships)
        
        # Increment correct soft relationship count if soft evaluation score is 1 (full match)
        if calculate_soft_evaluation_score(actual_relationships_json, predicted_relationships_json) == 1:
            correct_soft_relationships += 1

# Calculate accuracy for hard evaluation
accuracy_relationships = correct_relationships / total_relationships if total_relationships > 0 else 0

# Calculate accuracy for soft evaluation
accuracy_soft_relationships = correct_soft_relationships / total_predicted_relationships if total_predicted_relationships > 0 else 0

# Print results
print("Total relationships:", total_relationships)
print("Correct relationships (Hard Evaluation):", correct_relationships)
print("Accuracy (Hard Evaluation):", accuracy_relationships)

print("\nTotal predicted relationships:", total_predicted_relationships)
print("Correct relationships (Soft Evaluation):", correct_soft_relationships)
print("Accuracy (Soft Evaluation):", accuracy_soft_relationships)

# Print hard evaluation score
print("\nHard Evaluation Score:", hard_evaluation_score)

# Print soft evaluation score
print("Soft Evaluation Score:", soft_evaluation_score)
```

```python
import csv
import json

# Initialize variables
total_relationships = 0
correct_relationships = 0
modified_relationships = 0  # New variable to count modified relationships

# Open the CSV file and iterate over rows
with open('test_processedgpt4fewshot.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        actual_relationships_json = json.loads(row['actual'])
        predicted_relationships_json = json.loads(row['predictedgpt4fewshot'])
        
        # Ensure 'relationships' key exists in predicted JSON and add 'from' key if missing
        if 'relationships' not in predicted_relationships_json:
            predicted_relationships_json['relationships'] = [{'from': ''}]
        elif predicted_relationships_json['relationships'] and 'from' not in predicted_relationships_json['relationships'][0]:
            predicted_relationships_json['relationships'][0]['from'] = ''
            modified_relationships += 1
        
        # Ensure 'relationships' key exists in actual JSON and add 'from' key if missing
        if 'relationships' not in actual_relationships_json:
            actual_relationships_json['relationships'] = [{'from': ''}]
        elif actual_relationships_json['relationships'] and 'from' not in actual_relationships_json['relationships'][0]:
            actual_relationships_json['relationships'][0]['from'] = ''
            modified_relationships += 1
        
        # Extract relationships from actual JSON
        actual_relationships = []
        for relation_type in ['from', 'target', 'relation']:
            actual_relationships.extend([relation.get(relation_type, '') for relation in actual_relationships_json.get('relationships', [])])
        
        # Extract relationships from predicted GPT-4 JSON
        predicted_relationships = []
        for relation_type in ['from', 'target', 'relation']:
            predicted_relationships.extend([relation.get(relation_type, '') for relation in predicted_relationships_json.get('relationships', [])])
        
        # Check for correct predictions
        for relation in predicted_relationships:
            if relation in actual_relationships:
                correct_relationships += 1
        
        # Increment total relationships count
        total_relationships += len(actual_relationships)

# Calculate accuracy
accuracy_relationships = (correct_relationships - modified_relationships) / total_relationships if total_relationships > 0 else 0

# Print results
print("Total relationships:", total_relationships)
print("Correct relationships:", correct_relationships)
print("Modified relationships:", modified_relationships)  # Added for clarity
print("Accuracy:", accuracy_relationships)
```
